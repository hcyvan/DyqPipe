# DyqPipe Development Notebook

## Python Script
### dyq_generate_matrix.py: Genereate matrix

```
python dyq_generate_matrix.py -i tmp_data/lung_cancer/lung_N1.bed,tmp_data/lung_cancer/lung_N2.bed,tmp_data/lung_cancer/lung_N3.bed,tmp_data/lung_cancer/lung_T1.bed,tmp_data/lung_cancer/lung_T2.bed,tmp_data/lung_cancer/lung_T3.bed -c ./tmp_data/lung_cancer/index.bed.gz
```

### dyq_dmc_finder.py

```
python .\dyq_dmc_finder.py find -i .\tmp_data\lung_cancer\lung_matrix.5mc.bed -c lung_N1,lung_N2,lung_N3 -t lung_T1,lung_T2,lung_T3 -o .\tmp_data\lung_cancer\lung_dmc.bed
python .\dyq_dmc_finder.py filter -i .\tmp_data\lung_cancer\lung.dmc.bed -o .\tmp_data\lung_cancer\lung.dmc_1.2_0.05_.bed
```

### dyq_matrix_extract.py

```
python .\dyq_matrix_extract.py dmc -m .\tmp_data\lung_cancer\lung_matrix.5hmc.bed -i .\tmp_data\lung_cancer\lung.dhmc_1.2_0.05_.bed -o .\tmp_data\lung_cancer\lung.dhmc_1.2_0.05_sample_dhmc.bed
python .\dyq_matrix_extract.py dmc -m .\tmp_data\lung_cancer\lung_matrix.5mc.bed -i .\tmp_data\lung_cancer\lung.dhmc_1.2_0.05_.bed -o .\tmp_data\lung_cancer\lung.dhmc_1.2_0.05_sample_dmc.bed
```

### dyq_dmr_finder.py

```
python .\dyq_dmr_finder.py -i .\tmp_data\lung_cancer\lung.dhmc_1.2_0.05_.bed -o  .\tmp_data\lung_cancer\lung.dhmr_1.2_0.05_.bed -d 200
python .\dyq_dmr_finder.py -i .\tmp_data\lung_cancer\lung.dmc_1.2_0.05_.bed -o  .\tmp_data\lung_cancer\lung.dmr_1.2_0.05_.bed -d 200
```

### dyq_plot.py

```
python .\dyq_plot.py parser_cpg_cor -g lung_N1,lung_N2,lung_N3 -a .\tmp_data\lung_cancer\lung.dhmc_1.2_0.05_.dmc_1.2_0.05_sample_dmc.bed -b .\tmp_data\lung_cancer\lung.dhmc_1.2_0.05_.dmc_1.2_0.05_sample_dhmc.bed -o  .\tmp_data\lung_cancer\control.png
```

### dyq_generate_report.py

```
python ./dyq_generate_report.py -o ./tmp_data/Report
```

## R Script
